## **Conjunto de Mandelbrot**

Es el conjunto de números complejos c para los cuales la función $f_c(z) = z^2 + c$ no diverge cuando se itera desde z=0 hasta un numero suficientemente grande, es decir que la secuencia $f_c(0), f_c(f_c(0 )), ...,$ permanece acotado. 

Pongamos atención en una familia polinomios muy especial, los polinomios cuadráticos $$P_{c}(z)=z^{2}+c,  \quad z\in \mathbb{C}$$
donde $c\in \mathbb{C}$ es un parametro complejo. Luego $\jmath_{c}$ denota el cnjunto de Julia del polinomio $P_{c}(z)$. El único punto critico de  $P_{c}(z)$ es el origen $z=0$. Considerando los iterados $P_{c}(0)=c, \quad P_{c}^{2}(0)=c^{2}+c, \quad P_{c}^{3}(0)=\left ( c^{2}+c \right )^{2}+c,\quad P_{c}^{4}(0)=\left (\left ( c^{2}+c \right )^{2}+c  \right )^{2}+c,...$ hay dos casos que pueden ocurrir. Si los iterados $P_{c}^{k}(0)$ estan acotados, entonces el conjunto de Julia $\jmath_{c}$ es conexo. En el otro caso,  los iterados  $P_{c}^{k}(0)$ tiendn a $\infty $ entonces el conjunto de Julia $\jmath_{c}$ es totalmenMte disconexo.
El conjunto de Mandelbrot se $M$ es el conjunto de parametros valores de $c$ tales que los iterados  $P_{c}^{k}(0)$ estan acotados.  Luego $c  \in M$ si y solo si, $\jmath_{c}$ es conexo. 

Teorema. Un numero complejo $c$ pertenece al conjunto de Mandelbrot $M$ si  y solo si $\left |P_{c}^{n}(0)  \right |\leq 2$, para todo $n\geq 1$. El conjunto de Mandelbrot es un subconjunto compacto de el disco cerrado $\left \{ \left | c \right | \leq 2\right \}$. Por eso $\mathbb{C}\setminus M$ es concexo.

## **Conjunto de Julia**
En 1906, Pierre Fatou describió un ejemplo sorprendente: Para la función
$$z \mapsto  \frac{z^2}{\left (z^{2}+2  \right )}$$mostró que casi todas las órbitas bajo iteración convergen a cero, aunque hay un conjunto de puntos excepcionales, llamado el conjunto de  Cantor para la cual la órbita permanece acotada lejos de cero.

Defincion de El conjunto de Fatou y Julia. Sea $S$ una superficie compacta de Riemann, sea $f:S\rightarrow S$ una función holomorfica no constante, y sea  $f^{\circ n}:S\rightarrow S$  su $n$-campo iterado. El dominio de normalidad pasra las colecciones de iterados $\left \{ f^{\circ n}:S\rightarrow S \right \}$ es llamado el **Conjunto de Fatou** para $f$, y su complemento es llamado el **Conjunto de Julia**. 

Para cualquier punto $p_{0}\in S$, existe la siguiente dicotomia.
- Caso 1:si existe alguna vecindad $U$ de $p_{0}$ tal que la secuencia de iterados $\left \{ f^{\circ n}:S\rightarrow S \right \}$ restringida a $U$ forma una familia normal de funciones de $U$ a $S$, entonces decimos que $p_{0}$ pertenece a el **Conjunto de Fatou** de $f$.
- Caso 2: si no existe una vecindad entonces decimos que $p_{0}$ pertenece a el **Conjunto de Julia** de $f$.

Por lo anterior podemos ver que el **Conjunto de Julia** es un subconjunto cerrado  de $S$, mientras que su complemento, el **Conjunto de Fatou** es un subconjunto abierto. Luego un punto $p_{0}$ pertenece al  **Conjunto de Fatou**  si las dinamicas en la vecindad de $p_{0}$ muestran dependencia sensible sobre las condiciones iniciales, es decir que pequeños cambios en las condiciones inciales conducen a comportamientos diferentes despues de un numero grande de iteraciones.

```python
# cargar librerias para procesamiento de imagenes
from PIL import Image
   
#Nombre de la funcion:f_generadora_conjunto_de_Julia
#Parametros de la funcion:Real_C(parte real de C),
#Imaginaria_C(parte imaginaria de C)
#Ancho(ancho de la imagen)
#Alto(alto de la imagen)
#Aumento(aumento a la imagen)

    
def f_generadora_del_conjunto_de_Julia(Real_C,Imaginaria_C,Ancho,Alto,Aumento):
    #establecer el ancho, la altura y el zoom la imagen a crear
    w, h, zoom = Ancho,Alto,Aumento
           
    # crea una nueva imagen en modo RGB espacio de colores Rojo, Verde, Azul y selecciona el blanco de fondo
    bitmap = Image.new("RGB", (w, h), "white")
  
    # Asigna el almacenamiento de la imagen cargando los datos de los píxeles de la imagen.
    pix = bitmap.load()
     
    # Configuracion de variables de acuerdo a la 
    # Ecuacion elejida para crear el conjunto de Julia
    # parametro c de la ecuacion  $f_c(z) = z^2 + c$ 
    cX, cY = Real_C, Imaginaria_C
    # asignamos el orgien en (0,0)
    moveX, moveY = 0.0, 0.0
    #moveX, moveY = -1.0, 0.0
    # definimos el numero maximo de iteraciones de la funcion  $f_{c}^{k}(0)$
    # si la condicion para pertenecer  conjunto de Julia se cumple $\left |f_{c}^{n}(0)  \right |\leq 2$ no  se continuan las iteraciones 
    maxIter = 255
   
    # se  itera por cada una de las lineas verticales que forman la imagen 
    for x in range(w):
        # se  itera por cada elemento de una linea vertical que forma la imagen
        for y in range(h):
            # Se asigna una cordenada del plano complejo a la  image
            zx = 1.5*(x - w/2)/(0.5*zoom*w) + moveX
            zy = 1.0*(y - h/2)/(0.5*zoom*h) + moveY
            i = maxIter
            # se define la condicion necesacia y suficiente para que u n parametro c pertenezca al conjunto de Julia          
            #si despues de  un numero maximo de iteraciones 'maxIter'  la norma de |z|=|f_{c}^{n}(0)|< 2, 
            while zx*zx + zy*zy < 4 and i > 1:
                # se define la parte real del cuadrado de z y se suma el termino real del parametro c
                tmp = zx*zx - zy*zy + cX
                # se define la parte imaginaria del cuadrado de z y se suma el termino imaginario de c
                zy,zx = 2.0*zx*zy + cY, tmp
                # se actualiza el contador i+1
                i -= 1
  
            # convierte byte to RGB (3 bytes), un poco de magia  para conseguir bonitos colores
            pix[x,y] = (i << 21) + (i << 10) + i*8
  
    # muestra el conjuto de Julia creado 
    bitmap.show()
```

```python
#instalar librerias
# librerias para el procesamiento de imagenes
#pip install pillow

# libreria para operaciones  con matrices
#pip install numpy
#import numpy as np

# Codigo en Python para generar graficamente el conjunto de Mandelbrot  
# importa las librerias necesasrias
# librerias para el procesamiento de imagenes
import numpy as np
from PIL import Image
# librerias para el procesamiento de mamtrices
from numpy import complex, array
## librerias para la conversion de sistemas de color p.e RGB
import colorsys

def rgb_conv(i):
    color = 255 * array(colorsys.hsv_to_rgb(i / 255.0, 1.0, 0.5))
    return tuple(color.astype(int))

# funcion que define el conjunto de Mandelbrot
def mandelbrot(x, y):
    # se crea el complejo c  donde x es la parte real e y la imaginaria 
    c0 = complex(x, y)
    # define el inicio de la iteracion f_c^n(0), en el punto de origen,  unico punto de equilibrio. 
    c = 0
    # se inician las iteraciones de la fucion z=z^2+c
    for i in range(1, 1000):
        #condicion suficiente y necesaria para pertenecer alcomplemto del conjunto de Julia,  el conjnto de Fatou, 
        if abs(c) > 2:
            # retorna un color si no pertenece al conjunto de Mandelbrot
            return rgb_conv(i)
        # 
        c = c * c + c0
    # si el resultado de iterar la ecuación anterior esta acotado  en un disco compacto de radio 2
    # se retorna un color negro
    return (0, 0, 0)



# establece el ancho de salida de la imagen como  por ejemplo 1024 
def f_generadora_del_conjunto_de_Madelbrot(Ancho):
    WIDTH = Ancho

    # crea una nueva imagen en el espacio de color  RGB  de alto igual aun medio del ancho
    img = Image.new('RGB', (WIDTH, int(WIDTH / 2)))
    # cargando los datos de píxeles de la imagen
    pixels = img.load()
    
    # funcion que retorna una tupla de colores (tupla:conjunto de datos cuyo orden es inalterable)
    # como valor entero de un  espacio de color rojo-verde-azul RGB



    #itera por el rango del ancho de la imagen
    for x in range(img.size[0]):
  
    # muestra el progreso como procentajes 
    #print("%.2f %%" % (x / WIDTH * 100.0)) 
        #itera por el rango del alto de la imagen
        for y in range(img.size[1]):
        # cada pixel de la imagen es representada como  un elemenento del  plano complejo c=x+yi, al que se evalua la condicion de Mandelbrot
            pixels[x, y] = mandelbrot((x - (0.75 * WIDTH)) / (WIDTH / 4),
                                      (y - (WIDTH / 4)) / (WIDTH / 4))
        
# muestra el conjunto de mandelbrot creado 
    img.show()
```

```python
# funcion generadora del conjunto de mandelbrot parauna imagen de 1024 pixeles
f_generadora_del_conjunto_de_Madelbrot(1024)
```

```python
# Otra forma de general el Conjunto de Mandelbrot

#parametros: 
#eje_x: objeto tipo lista con el intervalo del eje X o eje real que se va a graficar por ejemplo [-2.0,1.0]
# eje_y: objeto tipo lista con el intervalo del eje y o eje imaginario que se va a graficar por ejemplo [-1.0,1.5]
# iteraciones_maximas:numero  de iteraciones maximas a realzar para evaluar la funcion z=z^2+c
#dimensiones_imagen: objeto tipo listoa con el ancho y alto de la imagen en pixeles por ejemplo [1244,1224]

def f_generadora_del_conjunto_de_Mandelbrot_otra_forma(eje_x,eje_y,iteraciones_maximas,dimensiones_imagen):

    from PIL import Image
    #Parametros  
    # area dibujada
    #Eje real del plano complejo
    xa = eje_x[0]
    xb = eje_x[1]
    #Eje imaginario del plano complejo
    ya = eje_y[0]
    yb = eje_y[1]
  
    # maximas iteraciones permitidas
    maxIt = iteraciones_maximas 
  
    # tamaño de la imagen
    imgx = dimensiones_imagen[0]
    imgy = dimensiones_imagen[1]
    #creamos una imagen en modo RGB de tamaño imgx por imgy
    image = Image.new("RGB", (imgx, imgy))
    # asigna a los elementos de la imagen(pixeles) un valor en el plano complejo  
    for y in range(imgy):
        zy = y * (yb - ya) / (imgy - 1)  + ya
        for x in range(imgx):
            zx = x * (xb - xa) / (imgx - 1)  + xa
            z = zx + zy * 1j
            c = z
            # se inician las iteraciones de la fucion z=z^2+c
            for i in range(maxIt):
                 #condicion suficiente y necesaria para pertenecer alconjunto de Mandelbrot
                if abs(z) > 2.0: break
                z = z * z + c
            # segun i (numero de iteraciones) se asigna el color en el punto x,y
            image.putpixel((x, y), (i % 4 * 64, i % 8 * 32, i % 16 * 16))
    # muestra el conjunto de mandelbrot creado   
    image.show()
```
